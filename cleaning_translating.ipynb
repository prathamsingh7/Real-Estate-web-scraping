```python
import pandas as pd
from googletrans import Translator
```

```python
translator = Translator()
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(5)
```

```python
df['Other information'][1][:3]
```

```python
df['Other information'] = df['Other information'].apply(lambda x: x[2:])
```

```python
df['Other information']
```

```python
df.head(5)
```

```python
df['Registration Date'] = pd.to_datetime(df['Registration Date'])
```

```python
df.head(5)
```

```python
df.to_csv('cleaned_data.csv', index=False)
```

```python
for col in df.columns:
    if col in ['Document type', 'Revenue Office', 'Buyer name', 'Seller name', 'Other information']:
        df[col] = df[col].astype(str)
        new_column = []
        for t in df[col]:
            text = str(t)
            if text.strip():
                try:
                    translation = translator.translate(text, src='mr', dest='en')
                    new_column.append(translation.text)
                except:
                    new_column.append("")
            else:
                new_column.append("")
        
        df[col] = new_column
```
